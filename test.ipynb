```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('data/new_df.csv')
test_df = pd.read_csv('data/test.csv')
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
# for training data
train_df['datetime'] = pd.to_datetime(train_df['datetime']) # convert into datetime and then extract the hour, month and year
train_df['hour'] = train_df['datetime'].dt.hour
train_df['day'] = train_df['datetime'].dt.day
train_df['month'] = train_df['datetime'].dt.month
train_df['year'] = train_df['datetime'].dt.year
```

### Removing unecessary columns

```python
train_df.drop(columns=['datetime', 'casual', 'registered'], axis=1, inplace=True)
```

```python
# for testing data
test_df['datetime'] = pd.to_datetime(test_df['datetime']) # convert into datetime and then extract the hour, month and year
test_df['hour'] = test_df['datetime'].dt.hour
test_df['day'] = test_df['datetime'].dt.day
test_df['year'] = test_df['datetime'].dt.year
test_df['month'] = test_df['datetime'].dt.month
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
test_df.drop(columns=['datetime'], axis=1, inplace=True)
```

### Seperate X and y

```python
# for training data
X = train_df.drop(columns=['count'])
y = train_df['count']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.95, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_y_pred = lr_model.predict(X_test)

# Evaluate
rmse = mean_squared_error(y_test, lr_y_pred)
r2 = r2_score(y_test, lr_y_pred)
print(rmse)
print(r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_y_pred = rf_model.predict(X_test)

# Evaluate
rmse = mean_squared_error(y_test, rf_y_pred)
r2 = r2_score(y_test, rf_y_pred)
print(rmse)
print(r2*100)
```
